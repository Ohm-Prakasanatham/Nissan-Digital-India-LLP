1. [Python] Given two lists of numbers, one with actual values and another one with predictions, implement a function to calculate the Root Mean Squared Error.

```python

def calculate_rmse(actual_value,predicted_values):
  mse = np.mean((np.array(actual_value) - np.array(predicted_values)) **2)
  rmse = np.sqrt(mse)
  return rmse

actual_value = [0,1,2,0,3]
predicted_values = [0.2, 2.3, 4.5, 0.5, 1.1]
rmse = calculate_rmse(actual_value,predicted_values)
print("rms error:",rmse)
```

2. [Pandas / Numpy] Subset the dataframe:

    a. New dataframe’s columns should contain first 5 columns and last 2 columns of the original dataframe

    b. New dataframe’s rows should contain first 10 rows and last 3 rows of the original dataframe

```python
subset_columns = df.iloc[:, :5].join(df.iloc[:, -2:])
```

```python
subset_rows = pd.concat([df.head(10), df.tail(3)])
```

```python
print("Subset based on columns:")
print(subset_columns)
print("\nSubset based on rows:")
print(subset_rows)
```
